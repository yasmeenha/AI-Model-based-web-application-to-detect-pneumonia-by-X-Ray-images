```python
import pandas as pd

columns = ['RI','Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba','Fe','Type']
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data', names=columns, header=None)
bins = [0.0, 0.8, 1.0, float('inf')]
df1 = pd.DataFrame(columns=['Mg1', 'Mg2', 'Mg3', 'Al', 'K', 'Type'])
df2 = pd.DataFrame(columns=['Mg1', 'Mg2', 'Mg3', 'Al', 'K', 'Type'])
df3 = pd.DataFrame(columns=['Mg1', 'Mg2', 'Mg3', 'Al', 'K', 'Type'])

v2_1 = df[(df['Mg'] >= bins[0]) & (df['Mg'] < bins[1])]
v2_2 = df[(df['Mg'] >= bins[1]) & (df['Mg'] < bins[2])]
v2_3 = df[(df['Mg'] >= bins[2])]
bin_range = bins[0] + (bins[1] - bins[0]) / 2.0
for index, row in df.iterrows():
    if row["Mg"] == "v2_1":
        v2_1_row = v2_1.loc[v2_1["id"] == row["id"]]
        if len(v2_1_row) == 1:
            df1.at[index, "Mg1"] = v2_1_row["Mg"].values[0]
        else:
            df1.at[index, "Mg1"] = row["Mg"]
    else:
        df1.at[index, "Mg1"] = row["Mg"]
df1['Mg2'] = bin_range
df1['Mg3'] = bin_range
df1['Al'] = df['Al']
df1['K'] = df['K']
df1['Type'] = df['Type'] 
print(df1)
for index, row in df.iterrows():
    if row["Mg"] == "v2_2":
        v2_2_row = v2_2.loc[v2_2["id"] == row["id"]]
        if len(v2_2_row) == 1:
            df2.at[index, "Mg2"] = v2_2_row["Mg"].values[0]
        else:
            df2.at[index, "Mg2"] = row["Mg"]
    else:
        df2.at[index, "Mg2"] = row["Mg"]
df2['Mg1'] = 0.9
df2['Mg3'] = 0.9
df2['Al'] = df['Al']
df2['K'] = df['K']
df2['Type'] = df['Type'] 
print(df2)
for index, row in df.iterrows():
    if row["Mg"] == "v3_2":
        v2_3_row = v2_3.loc[v2_3["id"] == row["id"]]
        if len(v2_3_row) == 1:
            df3.at[index, "Mg3"] = v2_3_row["Mg"].values[0]
        else:
            df3.at[index, "Mg3"] = row["Mg"]
    else:
        df3.at[index, "Mg3"] = row["Mg"]
df3['Mg1'] = 0.4
df3['Mg2'] = 0.4
df3['Al'] = df['Al']
df3['K'] = df['K']
df3['Type'] = df['Type'] 
print(df3)
```
